```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

scores = df[["math score", "reading score", "writing score"]]

# histogram
scores.hist(bins=10)

plt.suptitle("Histogram of Student Scores")
plt.show()

# boxplot
plt.boxplot(scores.values, labels=scores.columns)

plt.title("Boxplot of Student Scores")
plt.ylabel("Scores")

plt.show()
```
